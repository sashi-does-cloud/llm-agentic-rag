# LangChain + Google Gemini — Quick Start

This notebook demonstrates how to use **LangChain** with **Google Gemini (via `langchain-google-genai`)** to send structured chat messages and get AI responses.

---

### Prerequisites
- A **Google AI Studio** account with a Gemini API key
- Running on **Google Colab** (uses `google.colab.userdata` for secrets)

### What this notebook covers
1. Installing the required package
2. Initialising the Gemini chat model via LangChain
3. Building and invoking a message chain (System + Human messages)
4. Printing the model response

## 2. Import Libraries

| Import | Purpose |
|---|---|
| `init_chat_model` | Universal LangChain factory — picks the right provider from the model string |
| `userdata` | Colab secret manager — avoids hardcoding API keys in the notebook |
| `HumanMessage` | Represents a message from the user |
| `SystemMessage` | Sets the assistant's persona / behaviour instructions |
| `requests` | Standard HTTP library (imported for potential future use) |

```python
from langchain.chat_models import init_chat_model
from google.colab import userdata
from langchain.messages import HumanMessage, SystemMessage
import requests
```

## 3. Load API Key & Initialise the Model

**Model string format:** `"<provider>:<model-name>"`  
- Provider: `google_genai`  
- Model: `gemini-2.5-flash` (fast, cost-efficient; swap for `gemini-2.5-pro` for heavier tasks)

> `init_chat_model` auto-detects the provider from the prefix and returns a configured `ChatGoogleGenerativeAI` instance.

```python
api_key = userdata.get("GEMINI_API_KEY")

model = init_chat_model(
    "google_genai:gemini-2.5-flash",
    api_key=api_key,
)
```

## 4. Build the Message List

LangChain chat models accept a **list of message objects** that represent the conversation history.  
There are three main message types:

- **`SystemMessage`** — injected before the conversation; sets tone, rules, or persona for the model
- **`HumanMessage`** — a turn from the user
- **`AIMessage`** — a past response from the model (used when replaying history)

Ordering matters: `[system, human, ai, human, ...]` mirrors a real conversation.

```python
system_msg = SystemMessage("You are a helpful assistant.")
human_msg  = HumanMessage("What are AI Agents?")

messages = [system_msg, human_msg]
```

## 5. Invoke the Model & Print Response

`model.invoke(messages)` sends the full message list to the Gemini API and returns an **`AIMessage`** object.

Key attributes of the response object:
- `response.content` → the plain text answer *(most useful)*
- `response.response_metadata` → token counts, finish reason, model version
- `response.usage_metadata` → input / output token breakdown

> **Tip:** Use `print(response.content)` for cleaner output instead of printing the whole object.

```python
response = model.invoke(messages)

print(response)
```

## Next Steps

Here are some directions to explore from here:
